```python
from typing_extensions import NewType
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import Node2Vec
from torch_geometric.data import Data
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.svm import OneClassSVM
from PIL import Image, ImageEnhance, ImageFilter

from openood.evaluation_api import Evaluator
from openood.networks import ResNet18_32x32, ResNet18_224x224, ResNet50
from openood.networks.conf_branch_net import ConfBranchNet
from openood.networks.godin_net import GodinNet
from openood.networks.rot_net import RotNet
from openood.networks.csi_net import CSINet
from openood.networks.udg_net import UDGNet
from openood.networks.cider_net import CIDERNet
from openood.networks.npos_net import NPOSNet
from openood.networks.palm_net import PALMNet
from openood.networks.t2fnorm_net import T2FNormNet
from openood.postprocessors import BasePostprocessor # Import BasePostprocessor
from openood.utils import Config

# Data Corruption Functions (same as before)
def gaussian_noise(image, mean=0, std=25):
    noise = np.random.normal(mean, std, image.size)
    noisy_image = np.array(image) + noise
    return Image.fromarray(np.clip(noisy_image, 0, 255).astype(np.uint8))

def change_brightness(image, factor):
    enhancer = ImageEnhance.Brightness(image)
    return enhancer.enhance(factor)

def change_contrast(image, factor):
    enhancer = ImageEnhance.Contrast(image)
    return enhancer.enhance(factor)

def blur_image(image, radius):
    return image.filter(ImageFilter.GaussianBlur(radius))

def corrupt_image(image):
    corruptions = [
        lambda img: gaussian_noise(img),
        lambda img: change_brightness(img, np.random.uniform(0.5, 1.5)),
        lambda img: change_contrast(img, np.random.uniform(0.5, 1.5)),
        lambda img: blur_image(img, np.random.uniform(0, 2))
    ]
    corruption = np.random.choice(corruptions)
    return corruption(image)

# Feature Extraction using OpenOOD model
def extract_features(images, model, corrupt_prob=0.0):
    features = []
    model.eval()
    with torch.no_grad():
        for img in images:
            if np.random.random() < corrupt_prob:
                img = corrupt_image(img)
            img_tensor = model.preprocessing(img).unsqueeze(0)
            feat = model.feature_extractor(img_tensor)
            features.append(feat.squeeze().cpu().numpy())
    return np.array(features)

# Graph Construction and Embedding (same as before)
def create_graph(features, k=5):
    similarity_matrix = cosine_similarity(features)
    edges = []
    for i in range(len(features)):
        neighbors = similarity_matrix[i].argsort()[-k-1:-1]
        edges.extend([(i, j) for j in neighbors])

    edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
    return Data(x=torch.tensor(features, dtype=torch.float), edge_index=edge_index)

def embed_graph(graph, dim=64, walk_length=20, num_walks=10):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = Node2Vec(graph.edge_index, embedding_dim=dim, walk_length=walk_length,
                     context_size=10, walks_per_node=num_walks, num_negative_samples=1,
                     p=1, q=1, sparse=True).to(device)

    loader = model.loader(batch_size=128, shuffle=True, num_workers=4)
    optimizer = torch.optim.SparseAdam(list(model.parameters()), lr=0.01)

    model.train()
    for epoch in range(100):
        for pos_rw, neg_rw in loader:
            optimizer.zero_grad()
            loss = model.loss(pos_rw.to(device), neg_rw.to(device))
            loss.backward()
            optimizer.step()

    return model.embedding.weight.cpu().detach().numpy()

class GraphEmbeddingPostprocessor(BasePostprocessor):
    def __init__(self,config ,net,embed_dim=64):
        super().__init__(config)
        self.embed_dim = embed_dim
        self.ood_detector = OneClassSVM(kernel='rbf', nu=0.1)
        self.graph = None
        self.embeddings = None


    def setup(self, net, id_loader):
        # Extract features
        features = []
        labels = []
        net.eval()
        with torch.no_grad():
            for batch in id_loader:
                images, batch_labels = batch[0], batch[1]
                batch_features = net(images)
                features.append(batch_features.cpu().numpy())
                labels.extend(batch_labels.cpu().numpy())

        features = np.concatenate(features, axis=0)

        # Create and embed graph
        self.graph = self.create_graph(features)
        self.embeddings = self.embed_graph(self.graph)

        # Fit OOD detector
        self.ood_detector.fit(self.embeddings)

    def postprocess(self, net, data):
        net.eval()
        with torch.no_grad():
            features = net(data).cpu().numpy()

        # Add new nodes to the graph
        new_graph = self.update_graph(features)

        # Embed the updated graph
        all_embeddings = self.embed_graph(new_graph)

        # Get embeddings for the new data
        new_embeddings = all_embeddings[-len(features):]

        # Predict OOD scores
        ood_scores = self.ood_detector.score_samples(new_embeddings)

        return torch.from_numpy(-ood_scores)  # Negative because lower scores are more OOD

    def create_graph(self, features, k=5):
        similarity_matrix = cosine_similarity(features)
        edges = []
        for i in range(len(features)):
            neighbors = similarity_matrix[i].argsort()[-k-1:-1]
            edges.extend([(i, j) for j in neighbors])

        edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()
        return Data(x=torch.tensor(features, dtype=torch.float), edge_index=edge_index)

    def update_graph(self, new_features):
        if self.graph is None:
            return self.create_graph(new_features)

        old_features = self.graph.x.numpy()
        combined_features = np.vstack([old_features, new_features])
        return self.create_graph(combined_features)

    def embed_graph(self, graph, dim=64, walk_length=20, num_walks=10):
        device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        model = Node2Vec(graph.edge_index, embedding_dim=dim, walk_length=walk_length,
                         context_size=10, walks_per_node=num_walks, num_negative_samples=1,
                         p=1, q=1, sparse=True).to(device)

        loader = model.loader(batch_size=128, shuffle=True, num_workers=4)
        optimizer = torch.optim.SparseAdam(list(model.parameters()), lr=0.01)

        model.train()
        for epoch in range(100):
            for pos_rw, neg_rw in loader:
                optimizer.zero_grad()
                loss = model.loss(pos_rw.to(device), neg_rw.to(device))
                loss.backward()
                optimizer.step()

        return model.embedding.weight.cpu().detach().numpy()

# Main process
def main(args):
    # Initialize the base model
    if args.network == 'resnet18_32x32':
        base_model = ResNet18_32x32(num_classes=args.num_classes)
    elif args.network == 'resnet18_224x224':
        base_model = ResNet18_224x224(num_classes=args.num_classes)
    elif args.network == 'resnet50':
        base_model = ResNet50(num_classes=args.num_classes)
    else:
        raise ValueError(f"Unsupported network: {args.network}")

    config=Config()
     # Initialize our custom postprocessor
    postprocessor = GraphEmbeddingPostprocessor(config,base_model,embed_dim=64)

    # Initialize the OpenOOD evaluator
    evaluator = Evaluator(
        base_model,
        id_name=args.id_dataset,
        data_root=args.data_root,
        config_root=args.config_root,
        preprocessor=None,  # default preprocessing for the target ID dataset
        postprocessor_name='graph',  # we're using our custom postprocessor
        postprocessor=postprocessor,
        batch_size=args.batch_size,
        shuffle=False,
        num_workers=args.num_workers
    )

    # Run evaluation
    metrics = evaluator.eval_ood()
    print(metrics)

if __name__ == '__main__':
    import argparse
    args = argparse.Namespace()
    args.network = 'resnet18_32x32'
    args.num_classes = 10
    args.id_dataset = 'cifar10'
    args.data_root = './data'
    args.config_root = None
    args.checkpoint_path = None
    args.batch_size = 200
    args.num_workers = 2


    main(args)
```
